Так же, видно из полученных данных, что в них есть ошибки:

* минимальная высота потолков - `1` м.
* максимальная высота потолков - `100` м.

Выясним, какие строки содержат ошибки, и заполним их медианным значением.

```python
ceiling_height_meadian = df['ceiling_height'].median() # создаем переменную с медианным значением

df['ceiling_height'] = df['ceiling_height'].fillna(ceiling_height_meadian) # заполняем пропуски
```

Согласно СП 54.13330.2016:

* Высота (от пола до потолка) жилых комнат и кухни должна быть не менее `2,5` м.
* Высота внутриквартирных коридоров, холлов, передних, антресолей и под ними определяется условиями безопасности передвижения людей и должна составлять не менее `2,1` м.

Максимальная высота потолков не регламинтируется, но, вероятно, она не может быть более `5` м.

Исправим данные значения:

```python
df[(df['ceiling_height'] <= 2.1) | (df['ceiling_height'] >= 5)]['ceiling_height'].count()
```

Всего таких значений `51`, исправим:

```python
def change_ceiling(height):
    if height <= 2.1 or height >= 5:
        height = ceiling_height_meadian
    return height

df['ceiling_height'] = df['ceiling_height'].apply(change_ceiling)
```

Приступим к заполнению пустых значений площади кухни и жилой площади:

Чтобы заполнить эти пропуски, создадим новый столбец с типами общей площади квартиры, и найдем в этих типах соотношения площади кухни и жилой к общей площади. А в случае, если какое-то из значений есть, то найдем разницу между общей площадью и существующим значением.

```python
df['estate_area_types'] = pd.qcut(df['total_area'], q=6) # создаем столбец с разделением по 6 параметрам

df['estate_area_types'].value_counts()
```

```python
# создаем сводную таблицу по типу данных для заполнения

estate_area_types = df.pivot_table(index='estate_area_types',
                                   values=['total_area', 'living_area', 'kitchen_area'],
                                   aggfunc='mean')

# находим соотношения

estate_area_types['living_area_ratio'] =  estate_area_types['living_area'] / estate_area_types['total_area']
estate_area_types['kitchen_area_ratio'] =  estate_area_types['kitchen_area'] / estate_area_types['total_area']

estate_area_types
```

```python
# создаем функцию для заполнения пропущенных значений в площади кухни

def fill_kitchen_area(row):
    if pd.isna(row['kitchen_area']) and pd.isna(row['living_area']):
        return round((estate_area_types
                      .loc[row['estate_area_types']]['kitchen_area_ratio'] 
                      * row['total_area']), 2)
    
    # в случае, если площади кухни нет, а жилая площадь есть, то найдем разницу
    if pd.isna(row['kitchen_area']) and not pd.isna(row['living_area']):
        return row['total_area'] - row['living_area'] 
    return row['kitchen_area']
 
df['kitchen_area'] = df.apply(fill_kitchen_area, axis=1)
```

```python
# создаем функцию для заполнения пропущенных значений в жилой площади

def fill_living_area(row):
    if pd.isna(row['living_area']):
        return round((estate_area_types
                      .loc[row['estate_area_types']]['living_area_ratio'] 
                      * row['total_area']), 2)
    return row['living_area']
 
df['living_area'] = df.apply(fill_living_area, axis=1)
```

Создадим сводную таблицу по среднему расстоянию:

```python
# создаем сводную таблицу с средним расстоянием до аэропорта:

airports = df[~df['airports_nearest'].isna()].pivot_table(index='locality_name', values='airports_nearest', aggfunc='mean')

airports
```

Выясним, в сколько строк в пустыми значениями в расстоянии до аэропорта в известных нам городах:

```python
# делаем срез, где есть город в сводной таблице и нет расстояния до аэропорта

df.query('locality_name in @airports.index and airports_nearest.isna()')
```

Такие пропуски есть, значит их можно заполнить средним значением расстояния по городу:

```python
def fill_airports(row):
    if pd.isna(row['airports_nearest']) and row['locality_name'] in airports.index:
        return airports.loc[row['locality_name'], 'airports_nearest']
    return row['airports_nearest']
 
df['airports_nearest'] = df.apply(fill_airports, axis=1)
```

Всего строк с условием `133`, заполним их средним расстоянием по городу:

Остальные строки оставим без значений.

```python
def fill_centers(row):
    if pd.isna(row['city_centers_nearest']) and row['locality_name'] in locality_center_mean.index:
        return locality_center_mean.loc[row['locality_name'], 'city_centers_nearest']
    return row['city_centers_nearest']
 
df['city_centers_nearest'] = df.apply(fill_centers, axis=1)
```

### Изменение типов данных:

Заменим значения `float` на `int` там, где это уместно и сменим формат времени для дальнейшего исследования.

Изменяем значения в тех столбцах, в которых указано количество:

* количество этажей

**Изменим формат времени:**

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

**Заполним пропуски в столбце количества дней от публикации до снятия:**

Вероятно, что отсутствие значений в этом столбце указывает на то, что во время выгрузки данных, некоторые объявления о продаже не были сняты, соотвественно:

* можно оставить значения пустыми
* узнать дату выгрузки данных - найти самое последнюю дату и найти разницу между публикацией - столько дней это объявление находилось в продаже до выгрузки данных

Воспользуемся вторым методом и создадим новый столбец с количеством дней публикации.

```python
# выясняем дату выгрузки датасета

last_day = df['first_day_exposition'].max() 

last_day
```

```python
# создаем функцию по заполнению значений

def fill_days(row):
    if pd.isna(row['days_exposition']):
        return pd.Timedelta(last_day - row['first_day_exposition']).days # дельта времени по дням
    return row['days_exposition']
 
df['days_exposition_load'] = df.apply(fill_days, axis=1) # создаем новый столбец с временем публикации

df['days_exposition_load'] = df['days_exposition_load'].astype('int') # меняем на целое число
```
